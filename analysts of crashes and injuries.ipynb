```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv('crash_catalonia.csv')
data.describe()
```

```python
plt.figure( figsize=(6,6) )
x=DataFrame(data,columns=['Total_crashes'])
y=DataFrame(data,columns=['injuries'])
plt.scatter( x , y, alpha=0.3)
plt.title('Crashes and injuries during a week')
plt.xlabel('Number of Crashes')
plt.ylabel('injuries')
plt.xlim(0,25000)
plt.ylim(0,7000)
plt.show()
```
